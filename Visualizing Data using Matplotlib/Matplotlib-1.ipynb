### Restyle our line (color, line style, markers, width)

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y,label="2x",color="red",linewidth="5",linestyle="--",marker=".",markersize="20",markeredgecolor="blue")
plt.title("Our First Graph!", fontdict={"fontname":"Comic Sans MS","fontsize":"20"})
plt.xlabel("X Axis!", fontdict={"fontname":"Comic Sans MS","fontsize":"15"})
plt.ylabel("Y Axis!")
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,10])
plt.legend()
plt.show()
```

### Shorthand notation to restyle lines

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y, 'b^--', label="2x")
plt.title("Our First Graph!", fontdict={"fontname":"Comic Sans MS","fontsize":"20"})
plt.xlabel("X Axis!", fontdict={"fontname":"Comic Sans MS","fontsize":"15"})
plt.ylabel("Y Axis!")
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,10])
plt.legend()
plt.show()
```

### Plot more complex lines

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
x2 = np.arange(0,4.5,0.5)
print(x2)
# 1st Line
plt.plot(x, y, 'b^--', label="2x")
# 2nd line
plt.plot(x2,x2**2, "r-", label="x^2")
plt.plot()
plt.title("Our First Graph!", fontdict={"fontname":"Comic Sans MS","fontsize":"20"})
plt.xlabel("X Axis!", fontdict={"fontname":"Comic Sans MS","fontsize":"15"})
plt.ylabel("Y Axis!")
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,7,10])
plt.legend()
plt.show()
```

```python
# To show a projection of the 2nd line
x = [0,1,2,3,4]
y = [0,2,4,6,8]
x2 = np.arange(0,4.5,0.5)
print(x2)

# 1st Line
plt.plot(x, y, 'b^--', label="2x")

# 2nd line: we show a projection
plt.plot(x2[:6],x2[:6]**2, "r-")  # 0-5 solid line
plt.plot(x2[5:],x2[5:]**2, "r--",label="x^2") # 5-end dashed line

plt.title("Our First Graph!", fontdict={"fontname":"Comic Sans MS","fontsize":"20"})
plt.xlabel("X Axis!", fontdict={"fontname":"Comic Sans MS","fontsize":"15"})
plt.ylabel("Y Axis!")
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,7,10])
plt.legend()
plt.show()
```

### Resize Graph

```python
# To show a projection of the 2nd line
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize Graph
plt.figure(figsize=(5,3), dpi=125)
plt.plot(x, y, 'b^--', label="2x")

plt.show()
```

### Save Graph

```python
# To show a projection of the 2nd line
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize Graph
plt.figure(figsize=(5,3), dpi=100)
plt.plot(x, y, 'b^--', label="2x")

# Save Graoh
plt.savefig("mygraph.png", dpi=300)
plt.show()
```

### Bar Chart

```python
labels = ["A","B","C"]
values = [1, 4, 2]
bars = plt.bar(labels, values)
bars[0].set_hatch("/")
bars[1].set_hatch("O")
bars[2].set_hatch("*")

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]
patterns = ["/","O","*"]
bars = plt.bar(labels, values)
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.show()
```

```python
labels = ["A", "B", "C"]
values = [1, 4, 2]
patterns = ["/","O","*"]
bars = plt.bar(labels, values)
for bar in bars:
    bar.set_hatch(patterns.pop())
plt.show()
```
