### Line Graph Example (Plotting Data from CSV file)

```python
gas = pd.read_csv("gas_prices.csv")
gas.head(2)
```

```python
plt.plot(gas["Year"], gas.USA)
plt.show()
```

```python
plt.plot(gas["Year"],gas["USA"])
plt.plot(gas.Year,gas.Canada)
plt.plot(gas.Year,gas["South Korea"])
plt.show()
```

```python
# Cleaning the graph

plt.figure(figsize=(8,5))
plt.title("Gas Prices over Time (in USD)")
plt.plot(gas.Year,gas.USA, label="USA")
plt.plot(gas.Year,gas.Canada, label="Canada")
plt.plot(gas.Year,gas["South Korea"], label="South Korea")
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
# Cleaning the graph

plt.figure(figsize=(8,5))
plt.title("Gas Prices over Time (in USD)")
plt.plot(gas.Year,gas.USA, 'b.-', label="USA")
plt.plot(gas.Year,gas.Canada, 'r.-' ,label="Canada")
plt.plot(gas.Year,gas["South Korea"], 'g.-' ,label="South Korea")
plt.xticks(gas.Year[::3])
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.legend()
plt.show()
```

```python
# Cleaning the graph
plt.figure(figsize=(8,5))
plt.title("Gas Prices over Time (in USD)")

for country in gas.iloc[:,1:]:
    plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3])
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.title("Gas Prices over Time (in USD)")
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.xticks(gas.Year[::3])
plt.legend(loc=1)
plt.show()
```

```python
# Cleaning the graph
plt.figure(figsize=(8,5))
plt.title("Gas Prices over Time (in USD)")

countries_to_look_at = ["Australia", "USA", "Canada", "South Korea"]
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3])
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.legend()
plt.show()
```

```python
# Cleaning the graph
plt.figure(figsize=(8,5))
plt.title("Gas Prices over Time (in USD)", fontdict={"fontweight":"bold", "fontsize":18})
plt.plot(gas.Year,gas.USA, 'b.-', label="USA")
plt.plot(gas.Year,gas.Canada, 'r.-' ,label="Canada")
plt.plot(gas.Year,gas["South Korea"], 'g.-' ,label="South Korea")
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel("Year")
plt.ylabel("US Dollars")
plt.legend()
plt.savefig("Gas_price_figure.png", dpi=300)
plt.show()
```

### Pie Chart #1 (Counting data in CSV) - Visualizing Soccer Foot Preferences

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]
print(left)
print(right)
```

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]
plt.pie([left, right])
plt.show()
```

```python
# Adding title, labels and changing the colors
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]
labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']
plt.pie([left, right], labels=labels , colors=colors, autopct='%.2f %%')
plt.title("Foot Preference of FIFA Players")
plt.show()
```

### Pie Chart #2 (More advance Pandas Example) - Weight Distribution of FIFA Players

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >=175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight > 200].count()[0]
print(light,light_medium,medium,medium_heavy,heavy)
```

```python
plt.style.use("ggplot")  # changes the color of the pie graph
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
plt.pie(weights, labels=labels, autopct="%.2f %%")
plt.show()
```

```python
plt.style.use("ggplot")  # changes the color of the pie graph
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
plt.pie(weights, labels=labels, autopct="%.2f %%", pctdistance=1.5)
plt.show()
```

```python
plt.style.use("ggplot")  # changes the color of the pie graph
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.1,.1,.1,.1,.1) # five values corresponding to each weight
plt.pie(weights, labels=labels, autopct="%.2f %%", pctdistance=0.8, explode=explode)
plt.show()
```

```python
plt.style.use("ggplot")  # changes the color of the pie graph
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.2,0,0,.4) # five values corresponding to each weight
plt.pie(weights, labels=labels, autopct="%.2f %%", pctdistance=0.8, explode=explode)
plt.title("Weight Distribution of FIFA Players (in lbs)")
plt.show()
```

```python
# Styling

plt.style.use("default")
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club == "New England Revolution"]["Overall"]
labels = ["FC Barcelona", "Real Madrid", "NE Revolution"]

plt.figure(figsize=(5,8))
box = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth':7, 'color':'red'})

for b in box["boxes"]:
    # Set edge color of box
    b.set(color='blue', linewidth=4)
    # Set color inside of box
    b.set(facecolor="grey")

plt.title("Professional Soccer Team Comparison")
plt.ylabel("FIFA Overall Rating")
plt.show()
```
